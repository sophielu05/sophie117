```python
import numpy as np
import matplotlib.pyplot as plt

# 生成四組隨機數據
x1 = np.random.randn(10)
x2 = np.random.randn(10)
x3 = np.random.randn(10)
x4 = np.random.randn(10)

# 設置顏色、標記、線條粗細和標記大小
colors = ['r', 'g', 'b', 'purple']
markers = ['o', 's', 'd', '^']
line_widths = [1, 2, 3, 4]
marker_sizes = [5, 6, 7, 8]

# 創建四個子圖
fig, axs = plt.subplots(4, 1, figsize=(8, 12))

# 繪製每個子圖
for i, data in enumerate([x1, x2, x3, x4]):
    axs[i].plot(data, color=colors[i], marker=markers[i],
                linewidth=line_widths[i], markersize=marker_sizes[i])
    axs[i].set_title(f"Plot {i+1}")

# 調整子圖之間的間距
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 生成日期和數據
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))

# 計算每個欄位的平均值
mean_values = df.mean()

# 設定顏色和標籤
colors = ['blue', 'green', 'red', 'purple']
labels = mean_values.index  # 'A', 'B', 'C', 'D'

# 繪製長條圖
plt.bar(labels, mean_values, color=colors, width=0.5)

# 添加標題和標籤
plt.title("Average Value of Each Column")
plt.xlabel("Columns")
plt.ylabel("Average Value")

# 顯示圖形
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 生成日期和數據
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))

# 計算每個欄位的平均值
mean_values = df.mean()

# 確保平均值是正數，以免圓餅圖出現錯誤
mean_values = mean_values.abs()

# 設定顏色
colors = ['blue', 'green', 'red', 'purple']

# 繪製圓餅圖
plt.pie(mean_values, labels=mean_values.index, colors=colors, autopct='%1.1f%%', startangle=140)

# 添加標題
plt.title("Percentage of Average Values for Each Column")

# 顯示圖形
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))

# Calculate the mean of each column
mean_values = df.mean()

# Create the figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define positions for the bars
xpos = np.arange(len(mean_values.index))
ypos = np.zeros(len(mean_values.index))
zpos = np.zeros(len(mean_values.index))

# Plot the 3D bars
dx = 0.5 * np.ones_like(xpos)
dy = 0.5 * np.ones_like(ypos)
dz = mean_values.values

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average')

# Set labels and title
ax.set_xlabel("Columns")
ax.set_ylabel("Y")
ax.set_zlabel("Mean Values")
ax.set_title("3D Bar Chart of Mean Values")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 生成随机数据
data = np.random.rand(4)  # 生成 4 个随机数
labels = ['A', 'B', 'C', 'D']  # 设置标签

# 创建图形和坐标轴
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 定义条形图的位置
xpos = np.arange(len(labels))
ypos = np.zeros(len(labels))
zpos = np.zeros(len(labels))

# 绘制 3D 条形图
dx = 0.5 * np.ones_like(xpos)
dy = 0.5 * np.ones_like(ypos)
dz = data

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average')

# 设置标签和标题
ax.set_xlabel("类别")
ax.set_ylabel("Y")
ax.set_zlabel("数值")
ax.set_title("随机数据 3D 条形图")

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 設定圖表樣式
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'

# 生成隨機數據
num_features = 4  # 特徵數量，對應 A, B, C, D
data = np.random.rand(100, num_features)  # 100 個樣本，4 個特徵
df = pd.DataFrame(data, columns=list("ABCD"))  # 使用 list("ABCD") 設定欄位名稱

# 計算相關係數矩陣
corrMatrix = df.corr()

# 繪製熱圖
plt.figure(figsize=(10, 10))
sns.heatmap(corrMatrix, annot=True, cmap='coolwarm', center=0)
plt.title('隨機數據相關係數圖 (ABCD)')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
# 創建數據集
data = {
'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81,
79, 84, 77, 80],
'Method': ['A', 'A', 'A', 'A', 'A', 'A', # 教學方法 A
'B', 'B', 'B', 'B', 'B', 'B', # 教學方法 B
'C', 'C', 'C', 'C', 'C', 'C'] # 教學方法 C
}
# 將數據轉換為 DataFrame
df = pd.DataFrame(data)
# 使用 statsmodels 中的 ols 方法進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
# 輸出 ANOVA 表
print(anova_table)
```
